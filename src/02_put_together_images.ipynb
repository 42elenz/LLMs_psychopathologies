```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.gridspec import GridSpec
from pathlib import Path
import numpy as np

image_dir = Path("../outputs/figs/")
image_files = [
    "Experimental_design.png",
    "human_vs_ai_violin_video_7.png",
    "human_vs_ai_violin_video_8.png",
    "human_vs_ai_violin_video_9.png",
]

def to_float01(img):
    if img.dtype == np.uint8:
        return img.astype(np.float32) / 255.0
    return img

def rgba_to_rgb_on_white(img):
    img = to_float01(img)
    if img.ndim == 3 and img.shape[2] == 4:
        rgb = img[..., :3]
        a = img[..., 3:4]
        return rgb * a + (1 - a) * 1.0
    return img

# Load images
images = [rgba_to_rgb_on_white(mpimg.imread(image_dir / f)) for f in image_files]

fig = plt.figure(figsize=(15, 7), dpi=300)
fig.patch.set_facecolor("white")

# Create grid: A spans top row, B/C/D share bottom row
gs = GridSpec(2, 3, figure=fig, height_ratios=[0.7, 1], wspace=0.08, hspace=0.0)

# A spans the full top row
ax_a = fig.add_subplot(gs[0, 0])

# B, C, D in the bottom row
ax_b = fig.add_subplot(gs[0, 1])
ax_c = fig.add_subplot(gs[1, 0])
ax_d = fig.add_subplot(gs[1, 1])

axes = [ax_a, ax_b, ax_c, ax_d]
labels = list("abcd")

for i, (ax, img) in enumerate(zip(axes, images)):
    ax.imshow(img, interpolation="none")
    ax.set_axis_off()
    ax.set_frame_on(False)
    for sp in ax.spines.values():
        sp.set_visible(False)
    ax.patch.set_alpha(0)
    
    ax.text(-0.02, 1.02, labels[i], transform=ax.transAxes,
            fontsize=9, va='bottom', ha='right')

output_path = image_dir / "combined_figure.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight', pad_inches=0.02, facecolor='white')
plt.savefig(output_path.with_suffix('.pdf'), bbox_inches='tight', pad_inches=0.02, facecolor='white')
plt.show()
```
